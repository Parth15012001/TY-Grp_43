```
df=df.replace('?',math.nan)
df.isnull().sum()
```

```
df=df.dropna()
df.isnull().sum()
```

```
scaler = MinMaxScaler()
df[['age','fnlwgt','capital-gain','educational-num','capital-loss','hours-per-week']] = scaler.fit_transform(df[['age','fnlwgt','capital-gain','educational-num','capital-loss','hours-per-week']])
df
```

```
le = LabelEncoder()
df['workclass'] = le.fit_transform(df.workclass.values)
df['occupation'] = le.fit_transform(df.occupation.values)
df['education'] = le.fit_transform(df.education.values)
df['relationship'] = le.fit_transform(df.relationship.values)

# df['marital-status'] = le.fit_transform(df.marital-status.values)
le.fit(df['marital-status'])
df['marital-status'] = le.transform(df['marital-status'])
df
```

```
one_hot = pd.get_dummies(df['race'])
df = df.drop('race',axis = 1)
df = df.join(one_hot)
df
```

```
one_hot = pd.get_dummies(df['gender'])
df = df.drop('gender',axis = 1)
df = df.join(one_hot)
df  
```

```
def compute(data):
  d=dict()
  for i in data:
    if i in d:
      d[i]+=1
    else:
      d[i]=1
  print(d)
  maxk=-1
  maxval=-1
  for k,v in d.items():
    if(maxval<v):
      maxval=v
      maxk=k 
  return maxk 
```

```
print(df['native-country'].nunique())
MaxProportionKey=compute(df['native-country'])
print(MaxProportionKey)
```

```
df['native-country'][df['native-country'] != MaxProportionKey] = 0
df['native-country'][df['native-country'] == MaxProportionKey] = 1
print(df['native-country'].unique())
```
